**String operations**

La variable inicial no cambia, strings son inmutables

Literals

````python
myStringVar = "copo de nieve"
print(f"Primera letra {myStringVar[0]} - {myStringVar[-len(myStringVar)]}")
print(f"Última letra {myStringVar[-1]} - {myStringVar[len(myStringVar) - 1]}")
print(f"Cualquier letra {myStringVar[2]}")
````

Slicing 
```python
string[start:end:step]
```

````python
myStringVar = "copo de nieve"
print(myStringVar[::])
print(myStringVar[0::2])
print(myStringVar[-1::])
print(myStringVar[-len(myStringVar):-1:])
print(myStringVar[0:2])
print(myStringVar[::2])
````

````python
print(myStringVar.upper())
print(myStringVar.lower())
print(myStringVar)
````

Method replace
```python
str.replace(old, new, count)
```

count hace referencia a la cantidad de intercambios a realizar

````python
myNewStringVar = myStringVar.replace('nieve', 'azucar')
print(myNewStringVar)
print(myStringVar)
````
